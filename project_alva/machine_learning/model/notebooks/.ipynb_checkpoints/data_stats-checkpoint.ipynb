#### Data Preparation

```python
# Translate bulgarien construction type to english
mapping_type = {'Гредоред': 'Timbered', 'ЕПК': 'EPK', 'ПК': 'PK', 'Панел': 'Panel', 'Тухла': 'Brick'}
data['construction_type'].replace(mapping_type, inplace=True)

# Fix typo
data['district'].replace({'Gotse Gelchev': 'Gotse Delchev'}, inplace=True)

# Drop NaN values in price - can be changed for more columns and reset index
missing_price_init = sum(pd.isnull(data['price']))
row_count_init = len(data.index)
print("Missing prices in data: " + str(missing_price_init))
print("Percentage missing data: " + str(round(missing_price_init/row_count_init, 4)))

data = data.dropna(subset=['price'])
data = data.reset_index(drop=True)

# NaN values for each column
print("After we remove the values: ")
data.isnull().sum(axis = 0)
```

#### NOTE:
* The following step is not needed when we do data preparation for the model
* XGBoost deals with missing values implicitly

```python
# Most frequent values imputer to missing data
year_mode = int(data.year.mode())
floor_mode = int(data.floor.mode())
data['year'].replace({np.nan: year_mode}, inplace=True)
data['floor'].replace({np.nan: floor_mode}, inplace=True)

print(data.year.mode())
print(data.floor.mode())
print("missing data")

data.isnull().sum(axis = 0)
```

```python
# Map regions to data - needed for the follwing aggregation of data
data.loc[:, 'region'] = data.loc[:, 'district'].apply(
    lambda x: mapping.SOFIA_NEIGHBOURHOOD_TO_REGION_MAPPING[x]
)
print(data.loc[[1560]])
```

#### Filter Outliers

```python
# To be able to plot the data and generalize the model better we need to deal with outliers

print(data.shape)
print(data['price'].quantile(q=0.99))
print(data['price'].quantile(q=0.01))
print(data['price'].describe())

display(Markdown('---'))
filter_data = data[data.price < data['price'].quantile(q=0.99)]
filter_data = filter_data[filter_data.price > filter_data['price'].quantile(q=0.01)]
print(filter_data.shape)
print(filter_data['price'].quantile(q=0.99))
print(filter_data['price'].quantile(q=0.01))
print(filter_data['price'].describe())
```

#### Feature Scaling - target y 

*This part is about statistics and target variable transformation <br>
*There is a strong posibility that the y needs to be transformed because is not following a Normal Distribution <br>

article: https://medium.com/@kamskijohnm2m/ames-housing-price-prediction-complete-ml-project-with-python-2af595a749d6

```python
import seaborn as sns

# First we need to plot the target variable to confirm that we don't have normal distribution.
sns.distplot(filter_data.price)
plt.show()
```

```python
# Lets see the difference after transformation

from scipy import stats
%matplotlib inline

data_test = filter_data.copy()

figure = plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
stats.probplot(data_test.price, plot = plt)
plt.title("Actual SalePrice")
plt.subplot(1,2,2)
data_test.price = np.log(data_test.price)
stats.probplot(data_test.price, plot = plt)
plt.title("SalePrice after log transformation")
plt.show()
# plt.savefig(‘SalePrice.jpg’)
```

```python
log_data = data_test.copy()

sns.distplot(log_data.price)
plt.show()
```

```python
# Add correlation coefficient to plot

def corrfunc(x, y, **kws):
    r, _ = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate('r = {:.2f}'.format(r), xy=(0.1, 1.0), xycoords=ax.transAxes)

pair_plot = seaborn.pairplot(data, size=1.5)
pair_plot.map_lower(corrfunc)
pair_plot.map_upper(corrfunc)
```

```python
# Additional column park_district: True if there is a park in the district (or near the border of it)
# False otherwise

data.loc[:, 'park_district'] = data.loc[:, 'district'].apply(
    lambda x: mapping.SOFIA_NEIGHBOURHOOD_PARK_TRUE_FALSE_MAPPING[x]
)
print(data.loc[[1560]])
data.shape
```

```python
# Additional column for the population adjusted with +10%

data.loc[:, 'population'] = data.loc[:, 'region'].apply(
    lambda x: mapping.SOFIA_REGIONS_POPULATION_MAPPING[x]
)
data['adj_population'] = data.population.map(lambda x: (x + x*0.1))

print(data.loc[[1560]])
data.shape
```
